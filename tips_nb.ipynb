```python
X=df[['tip', 'sex','day', 'time', 'size', 'total_bill']]
y=df['smoker']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_test['sex'] = le1.transform(X_test['sex'])

y_train = le2.fit_transform(y_train)
y_test = le2.transform(y_test)

X_train['time'] = le3.fit_transform(X_train['time'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
X_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'), [3])], remainder='passthrough')
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
day_ct = ColumnTransformer(
	transformers=[('day_onehot', OneHotEncoder(handle_unknown='ignore'), [3])],
	remainder='passthrough'
)

X_train = day_ct.fit_transform(X_train)
X_test = day_ct.transform(X_test)

if hasattr(X_train, "toarray"):
	X_train = X_train.toarray()
	X_test = X_test.toarray()

gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix, classification_report
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
